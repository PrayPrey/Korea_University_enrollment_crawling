```python
import pandas as pd

data = pd.read_csv('project_1_2.csv',encoding='utf-8-sig')

for i in range(len(data)):
    if type(data['place'][i]) == float:
        data['place'][i] = '미정'
        
data['place'][56]
data
```

```python
list_all = []

for i in data['place']:
    list_all.append(i.split())
list_all
```

```python
import re
import re

pattern = re.compile('[0-9]')
count = 1
list_a = []

for i in list_all:
    list_b = []
    for j in i:
        a= re.sub(pattern,'',j)

        list_b.append(re.sub('[()\-호]','',a).strip())
    list_a.append(list_b)
        

    count += 1
print(count)
        
list_a      
```

```python
import re

regex1 = re.compile(r'\w\D\d\D{1,2}')
regex2 = re.compile(r'\w\D\d\D\d\D')
count = 0
list_day = []

for i in list_all:
    list_b = []
    for j in i:
        match1 = re.search(regex1,j)
        match2 = re.search(regex2,j)
        try:
            if match1.group()[-1] == '-':
                list_b.append(match2.group())
            else:
                list_b.append(match1.group())
        except AttributeError:
            pass        
    list_day.append(list_b)
    count += 1
print(count)

for i in range(len(list_day)):
    if len(list_day[i]) == 0:
        list_day[i] =['미정']
list_day

# pd.DataFrame(list_day)
```

```python
list_day_2 = []
for i in list_day:
    list_day_2.append(' '.join(i))
list_day_2
```

```python
data_2 = pd.concat([data,pd.DataFrame(list_day)],axis=1)
string_1 = 'entire	campus	number	class_	seperate	major	class_name	prof	time	place	day1'.split('\t')
string_1.append('day2')
data_2.columns = string_1
data_2
```

```python
import re

regex = re.compile('\W')
regex_2 = re.compile('[0-9]')
pattern = re.compile('[0-9]{,3}호')


list_place = []
count=0
for i in list_all:
    list_b = []
    for j in i:
        a_0 = re.sub(pattern,'',j)
        a = re.sub(regex,'',a_0)
        if len(re.sub(regex_2,'',a).strip()) == 1:
            list_b.append('')
        else:
            list_b.append(re.sub(regex_2,'',a))
    list_place.append(list_b)
    count+=1
        


list_place_2 = []
for i in list_place:
    a = ' '.join(i).split()
    if len(a) == 0:
        list_place_2.append(['미정'])
    else:
        list_place_2.append(a)
        
list_place_2
# for i in range(len(list_place)):
#     if len(list_day[i]) == 0:
#         list_day[i] =['미정']
# list_day
    
```

```python
# 노가다로 제거.
# 마지막 제거과정 이 부분이 향후에도 노가다로 바뀔 것.
regex = re.compile('[A-Z]호')
list_place_3 = []
for i in list_place_2:
    list_b = []
    for j in i:
        if len(re.sub(regex,'',j).replace('지하','').replace('제강의실','').replace('SUPEX','').replace('HALL','')) ==0:
            pass
        else:
            a = re.sub(regex,'',j).replace('지하','').replace('제강의실','').replace('호SUPEX','').replace('HALL','')
            b =a.replace('LP','LG포스코경영관').replace('과도관','과학도서관').replace('교육관','운초우선교육관').replace(
                '미래융합','미래융합관').replace('사대','사범대학').replace('B','').replace('전산실습실','').replace(
                '실험실','').replace('우정간학관','우정간호학관').replace('정보통신관','우정정보관').replace(
                '현차관','현대자동차경영관')
            list_b.append(b)
    list_place_3.append(list_b)
    
list_place_3
```
